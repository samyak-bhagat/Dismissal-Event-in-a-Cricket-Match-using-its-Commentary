## Dismissal Events of Cricket Match

```python
import pandas
import matplotlib.pyplot as plt
```

```python
data = pandas.read_csv(r'C:\Users\Vanshika Sharma\Documents\Cognitive Dashboard\Scraped Matches\test_match.csv')
data
```

```python
''' Reverse the data (get commentary starting from the first ball) '''
data = data.reindex(index=data.index[::-1])
data = data.reset_index()
data = data.drop(columns=['index'])
data
```

```python
max_ball = data['Balls'][len(data)-1]
max_ball
```

```python
'''Dismissal Scorecard as a Dataframe.'''
dismissal_data = pandas.DataFrame(
                columns=['Batsman', 'Bowler', 
                         'Runs','Type of Dismissal', 
                         'Extra Man','Overs'])
```

```python
def calculate_runs(runs):
    if(runs == 'FOUR' or runs == 'FOUR.'):
        return 4
    elif(runs == 'SIX'):
        return 6
    elif(runs == '1'):
        return 1
    elif(runs == '3'):
        return 3
    elif(runs == '2'):
        return 2
    elif(runs == 'out' or runs == 'byes'
         or runs == 'leg' or runs == 'no'
         or runs == '5'):
        return 0
    else:
        return 0
```

```python
''' Extracting commentary only for dismissals'''

# Dictionary to store the runs scored by batsman 
batsman_dict = {} 
runs=0
for i in range(len(data['Commentry'])):    
    comm = data['Commentry'][i]
    
    # getting bowler's and batsman's name: by a set pattern in commentry
    bowler = comm.split(',')[0].split('to')[0].strip()
    batsman = comm.split(',')[0].split('to')[1].strip()
        
    # action taken on the ball i.e number of runs or is it a out
    action = (comm.split(',' or '.')[1]).split('!' or '.')[0].strip()
    action = action.split(' ')
    
    runs = calculate_runs(action[0])
   
    # adding batsman as a key in dictionary
    # and runs as its value
    if (batsman not in batsman_dict.keys()) and runs!=None:
        batsman_dict[batsman] = runs
    else:
        batsman_dict[batsman] += runs
        
    #print(batsman_dict)
    
    if action[0] == 'out':
        over = data['Balls'][i]
        out_type = action[1]
        
        if action[1]!='Lbw':
            extra_man = action[3]
            
            # add dismissal details in dataframe
            dismissal_data = dismissal_data.append({'Batsman':batsman, 'Bowler':bowler,
                            'Runs':batsman_dict[batsman], 'Type of Dismissal':out_type,
                            'Extra Man':extra_man, 'Overs':over}, ignore_index=True)
            runs=0
        else:
            # Lbw does not require an extra man 
            dismissal_data = dismissal_data.append({'Batsman':batsman, 'Bowler':bowler,
                            'Runs':batsman_dict[batsman],'Type of Dismissal':out_type,
                            'Extra Man':"-", 'Overs':over}, ignore_index=True)
            runs=0
```

```python
dismissal_data
```

```python
'''list of overs (divided in the block of 10)'''

# the maximum number of balls played in the match
max_ball = data['Balls'][len(data)-1]
over_axis=[]
for i in range(0,int(max_ball)+10,10):
    over_axis.append(i//10)
max_ball
```

```python
over_axis
```

```python
''' visualization '''
plt.figure(figsize=(13,5))
plt.grid()

plt.plot(over_axis, dismissal_data['Overs'], marker='o', ms=9,linewidth=3, color='black')
plt.bar(over_axis, dismissal_data['Runs'], color='grey' , ls='-')

plt.xticks(over_axis)
plt.title("Overs and Fall of Wickets")
plt.xlabel("Number of Wickets")
plt.ylabel("Overs and Runs Scored by the Batsman")
plt.legend("")
plt.show()
```

```python
''' Frequency of wickets by Bowlers '''

bowler_dict={}
for bowler in dismissal_data['Bowler']:
    if bowler not in bowler_dict.keys():
        bowler_dict[bowler] = 1
    else:
        bowler_dict[bowler] += 1
```

```python
''' Visualizing Wickets per Bowler in the Match'''
plt.bar(*zip(*bowler_dict.items()), color="grey",  edgecolor='black')

plt.title("Wicket per Bowler")
plt.ylabel("wickets")
plt.xlabel("Bowlers")
plt.show()
```
